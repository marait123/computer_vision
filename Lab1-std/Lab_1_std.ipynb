```python

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.
    # images[0] will be drawn with the title titles[0] if exists
    # You aren't required to understand this function, use it as-is.
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
```

```python
'''
Requirement 3 (Histogram):

1- For the given images ( in histogram folder): 
    Read the image.
    Apply histogram and show it.
Hint
    A) Use histogram (image) to get histogram. Try different values for nbins (256,64,8), What does it mean? {{nbins in this case means the number of pixel groups on the x-axis for example when it is = 8 it means that every (256/8) pixels are grouped together and any pixel that fall into any of these groups increases the count associated with that x-axis devision. }}
    B) and function(showHist) to draw it.
** 2- Draw a grey-scale image that has uniform histogram 
same number of pixels for all intensity levels) using code only. Let the size of the image be 256x256.
use np.ones to draw image with ones.
'''
def showHist2(img,histogramImg, title="",xlabel="",ylabel=""):
    plt.figure()
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    bar(histogramImg[1]*255, histogramImg[0], width=0.8, align='center')

images_paths = ['histogram/ex1.jpg','histogram/ex2.png','histogram/ex3.png']
images = []
# 1. read images
for image_path in images_paths:
    ex1 = io.imread(image_path, as_gray=True)   
    """
    A. Answer
    {{nbins in this case means the number of pixel groups on the x-axis for example when it is = 8 it means that every (256/8) pixels are grouped together and any pixel that fall into any of these groups increases the count associated with that x-axis devision. }}
    """
    hist_bins = [256,64,8]
    for bin_width in hist_bins:
        hist1 = histogram(ex1, nbins=bin_width)
        # B.    
        showHist2(ex1,hist1,"histogram of image: {} with bins={}".format(image_path,bin_width))
```

```python
# second version using showHist function

# 2. Draw a grey-scale image that has uniform histogram 
# plt.figure()
uniform_image = np.ones((256,256))
uniform_image[:,0] = 256
for i in range(256):
    uniform_image[:,i] *= (i/255)

plt.title("uniform image")
io.imshow(uniform_image)
uniform_hist = histogram(uniform_image, nbins=256)
# B.
# plt.xlabel("pixel values")
# plt.ylabel("number of occurrence of pixel")
showHist2(uniform_image,uniform_hist,"histogram of uniform image", 'pixel values', 'number of occurence')

```
